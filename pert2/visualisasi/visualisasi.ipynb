```python
print("Latihan 2: Permen\n")
print("Saya memilih menggunakan Pie chart karena cocok untuk menunjukkan komposisi atau persentase dari sebuah keseluruhan. Pada kasus ini, kita ingin menunjukkan persentase peluang memilih permen Snickers dibandingkan dengan permen lainnya dalam sekantong permen. Dengan menggunakan pie chart, kita dapat dengan jelas melihat kontribusi relatif dari setiap jenis permen terhadap total jumlah permen, sehingga memudahkan pemahaman tentang peluang memilih permen Snickers secara visual.\n")

import matplotlib.pyplot as plt

# Problem
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]

# Menghitung total jumlah permen
total_candies = sum(candy_counts)

# Menghitung persentase peluang memilih permen Snickers
snickers_index = candy_names.index('Snickers')
snickers_probability = candy_counts[snickers_index] / total_candies * 100

# Menampilkan persentase peluang memilih permen Snickers
print(f"Persentase kemungkinan memilih permen Snickers: {snickers_probability:.2f}%")

# Membuat diagram pie
plt.figure(figsize=(8, 8))
plt.pie(candy_counts, labels=candy_names, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Peluang Memilih Permen Snickers')
plt.axis('equal')
plt.show()

```

```python
print("Latihan 3: Makanan\n")
# Jawaban
print("Untuk kasus ini, saya menggunakan diagram batang horizontal. Alasan pemilihan diagram batang horizontal adalah sebagai berikut:\n")

print("1. Memudahkan Perbandingan: Diagram batang horizontal memudahkan untuk membandingkan jumlah penjualan dari masing-masing makanan penutup karena panjang batangnya berbanding lurus dengan jumlah penjualan.")

print("2. Menyoroti Makanan Penutup yang Tidak Populer: Dalam konteks ini, kita ingin menyoroti tiga makanan penutup yang paling tidak populer. Dengan menggunakan diagram batang horizontal, makanan penutup yang memiliki jumlah penjualan terendah akan terletak di bagian paling kiri grafik, sehingga lebih mudah diidentifikasi.")

print("3. Penyajian Data yang Jelas: Diagram batang horizontal memberikan tampilan yang jelas dan mudah dipahami bagi pembaca. Pembaca dapat dengan cepat melihat perbandingan antara makanan penutup yang satu dengan yang lainnya.\n")

print("Dengan menggunakan diagram batang horizontal, kita dapat dengan efektif menunjukkan popularitas relatif dari masing-masing makanan penutup, sambil menyoroti makanan penutup yang kurang diminati untuk dipertimbangkan dalam penghapusan dari menu restoran.\n")


import matplotlib.pyplot as plt

# Problem
dessert_sales = {
  'Lava Cake': 14,
  'Mousse': 5,
  'Chocolate Cake': 12,
  'Ice Cream': 19,
  'Truffles': 6,
  'Brownie': 8,
  'Chocolate Chip Cookie': 12,
  'Chocolate Pudding': 9,
  'Souffle': 10,
  'Chocolate Cheesecake': 17,
  'Chocolate Chips': 2,
  'Fudge': 9,
  'Mochi': 13,
}

# Sorting dessert_sales based on their sales count
sorted_dessert_sales = sorted(dessert_sales.items(), key=lambda x: x[1])

# Extracting the names of the least popular desserts
least_popular_desserts = [dessert[0] for dessert in sorted_dessert_sales[:3]]

# Displaying the least popular desserts
print("Makanan penutup yang disarankan untuk dikeluarkan dari menu:")
for dessert in least_popular_desserts:
    print("-", dessert)

# Plotting the dessert sales
dessert_names = [dessert[0] for dessert in sorted_dessert_sales]
dessert_sales_count = [dessert[1] for dessert in sorted_dessert_sales]

plt.figure(figsize=(10, 6))
plt.barh(dessert_names, dessert_sales_count, color='skyblue')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Nama Makanan Penutup')
plt.title('Popularitas Relatif Makanan Penutup')
plt.gca().invert_yaxis()  # Invert y-axis to display the most popular dessert at the top
plt.show()

```
